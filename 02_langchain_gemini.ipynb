이번에는 init_chat_model()를 이용하지 않고 (이와 같은 함수를 팩토리 함수라고 한다) 다이렉트로 llm 을 불러와 사용해 보도록 한다 

```python
#필요한 모듈은 비슷하나 팩토리 함수는 안불러와도 된다
#대신 google 생성형 ai 래퍼를 직접 불러온다 GoogleGenerativeAI
from langchain_core.prompts import ChatPromptTemplate    # 랭체인에서 제공하는 대화방식의 프롬프트 템플릿 
from langchain_google_genai import GoogleGenerativeAI    # 랭체인에서 제공하는 구글 생성형 ai 래퍼
from dotenv import load_dotenv                           # .env 파일에서 환경변수를 로드하는 함수
import os                                                # 운영체제 관련 기능을 제공하는 모듈


from langchain_core.output_parsers import StrOutputParser   # 랭체인에서 제공하는 문자열  출력


load_dotenv()                                           # .env 파일에서 환경변수를 로드
```

```python
# 자 그럼 바로 다이렉트로 GoogleGenerativeAI 클래스를 사용해 보자 

llm_model =  GoogleGenerativeAI (model= "gemini-3-flash-preview", temperature=1.0)

response =llm_model.invoke("레데리2 에서의 시대의 돈과 현재 시점의 돈 환율 알려줘")       # 직접 질문을 넣어서 호출
print(response)        # 내용을 출력  
```

```python
# 이제 프롬프츠 테플릿을 사용해 보자 파싱까지 해본다 

# llm 모델 선택은 똑같다 
llm_model = GoogleGenerativeAI(model= "gemini-3-flash-preview", temperature=1.0)
# 프롬프트 템플릿을 만든다 
prompt_template = ChatPromptTemplate.from_template("레데리2 에서의 시대의 돈과 현재 시점의 돈 환율 알려줘")

# 이제 체인을 만들어 준다 
chain = prompt_template | llm_model | StrOutputParser()          

response2 = chain.invoke({})

print(response2)
```

```python
# 이제 단순한 from_template 방식이 아닌 메세지 형식의 프롬프트 템플릿을 사용해 보자 

# llm 모델 선택은 똑같다 
llm_model = GoogleGenerativeAI(model= "gemini-3-flash-preview", temperature=1.0)

# from_templates 이 아니라 from_messages 방식을 사용해서 프롬프트 템플릿을 ㅁ나든다 
# 약간 틀리지만 system, user, assistant 메세지 형식을 지원한다 


prompt = ChatPromptTemplate.from_messages([
    ("system", "너는 미디어 영상 포트폴리오 제작 전문가야"),                      # system 메세지 
    ("human","미디어 포토폴리오 제작에 대해 알려줘"),                             # human 메세지
    ("ai", "미디어 영상 포토폴리오 제작은..."),                                  # ai 메세지 
    ("human","미디어 영상 포토폴리오 제작에 대해서 설명해줘")                      # human 메세지 
])

# 이제 체인을 만들어 준다 
chain = prompt | llm_model | StrOutputParser()   

response3 = chain.invoke({})     # 프롬프트에 변수가 없으므로 빈 딕셔너리 전달 
print(response3)                 #  결과물 출력

# 굙하물을 보려면 윗단계의 실행 결과를 clear 해 줘야 한다 (주피터스 노트북 특징 )
```
